```python
%%time
conn = sqlite3.connect('ServiceRequest.db') #connect to SQLite database
cursor = conn.cursor()
cursor.execute("DROP TABLE IF EXISTS servise_info;")

cursor.execute('''CREATE TABLE IF NOT EXISTS service_info (
                    unique_key INTEGER,
                    created_date NUMERIC,
                    closed_date NUMERIC,
                    agency TEXT,
                    agency_name TEXT,
                    complaint_type TEXT,
                    descriptor TEXT, 
                    location_type TEXT,
                    incident_zip REAL,
                    incident_address TEXT,
                    street_name TEXT, 
                    cross_street1 TEXT, 
                    cross_street2 TEXT, 
                    intersection_street1 TEXT,
                    intersection_street2 TEXT,
                    address_type TEXT,
                    city TEXT, 
                    landmark TEXT,
                    facility_type TEXT, 
                    status TEXT,
                    due_date NUMERIC, 
                    resolution_description TEXT,
                    resolution_action_updated_date NUMERIC, 
                    community_board TEXT, 
                    bbl REAL, 
                    borough TEXT, 
                    x_coordinate_state_plane REAL, 
                    y_coordinate_state_plane REAL, 
                    open_data_channel_type TEXT, 
                    park_facility_name TEXT, 
                    park_borough TEXT, 
                    vehicle_type TEXT,
                    taxi_company_borough TEXT, 
                    taxi_pick_up_location TEXT, 
                    bridge_highway_name TEXT, 
                    bridge_highway_direction TEXT, 
                    road_ramp TEXT,
                    bridge_highway_segment TEXT, 
                    latitude REAL, 
                    longitude REAL, 
                    location TEXT
                )''')


cols = ['Unique Key', 'Created Date', 'Closed Date', 'Agency', 'Agency Name', 'Complaint Type', 'Descriptor', 'Location Type', 'Incident Zip', 'Incident Address', 'Street Name', 'Cross Street 1', 'Cross Street 2', 'Intersection Street 1', 'Intersection Street 2', 'Address Type', 'City', 'Landmark', 'Facility Type', 'Status', 'Due Date', 'Resolution Description', 'Resolution Action Updated Date', 'Community Board', 'BBL', 'Borough', 'X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Open Data Channel Type', 'Park Facility Name', 'Park Borough', 'Vehicle Type', 'Taxi Company Borough', 'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp', 'Bridge Highway Segment', 'Latitude', 'Longitude', 'Location']
chunksize = 10000

# parse csv file and populate db in chunks
with open("311_Service_Requests_from_2010_to_Present_20240331.csv") as f:
    reader = csv.DictReader(f)

    chunk = []
    # for i, row in enumerate(reader, start=1): 
    for row in reader: 

        items = itemgetter(*cols)(row)
        chunk.append(items)

        if len(chunk) == chunksize:
            conn.executemany(
                """
                INSERT INTO service_info
                    VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
                """, chunk
            )
            chunk = []
        
    if chunk:
        conn.executemany(
            """
            INSERT INTO service_info
                VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)
            """, chunk
        )

conn.commit()
conn.close()
# time: 20min32.0s
```

```python
conn = sqlite3.connect('ServiceRequest.db') #connect to SQLite database
cursor = conn.cursor()
```

```python
# checking if there is indexing on the 'complaint_type' column
qr = cursor.execute("""SELECT name, sql
                    FROM sqlite_master
                    WHERE type='index' AND tbl_name='service_info' AND sql LIKE '%complaint_type%';
               """)
```

```python
qr.fetchall()
```

```python
qr2 = cursor.execute("CREATE INDEX idx_complaint_type ON service_info (complaint_type);")
```

```python
%%time
res_eff = cursor.execute("""SELECT complaint_type, COUNT(complaint_type) AS amount FROM service_info
                        GROUP BY complaint_type
                        ORDER BY amount DESC;
                    """)
# CPU times: total: 5.14 s
# Wall time: 5.65 s

#from CPU 1min 5s --> 5.14s
```

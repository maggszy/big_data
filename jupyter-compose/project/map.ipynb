# Agregacja

```python
n = 0.01
df['lat'] = n * (df['tweet_latitude'] // n) + 0.5*n
df['long'] = n * (df['tweet_longitude'] // n) + 0.5*n
df
```

```python
df_agg = df.groupby(['lat', 'long'], as_index=False).count()[['lat', 'long', 'tweet_latitude']]
df_agg.columns = ['lat', 'long', 'size']
#df_agg['size'] = np.log(df_agg['size'] + 10)
df_agg.sort_values('size', ascending=False).iloc[0:30]
```

```python
df_agg = df.groupby(['lat', 'long'], as_index=False).count()[['lat', 'long', 'tweet_latitude']]
df_agg.columns = ['lat', 'long', 'count']
df_agg['size'] = np.minimum(df_agg['count'], 1000) + 10
df_agg['color'] = np.minimum(df_agg['count'], 500)
df_agg
```

```python
# kolor i rozmiar
fig = px.scatter_mapbox(df_agg, lat="lat", lon="long", color='color', size='size', size_max=10, opacity=0.5, color_continuous_scale=px.colors.diverging.RdYlBu[::-1])
fig.write_html("mapa_agg.html")
```
